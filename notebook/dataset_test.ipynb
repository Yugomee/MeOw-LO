```python
import cv2
import matplotlib.pyplot as plt

test_image = cv2.imread(os.path.join(base_path, '00000001_000.jpg'))
test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)

pd_frame = pd.read_csv(os.path.join(base_path, '00000001_000.jpg.cat'), sep =' ', header=None)
landmarks = (pd_frame.to_numpy()[0][1:-1]).reshape((-1, 2))

for lmk in landmarks:
    x, y = int(lmk[0]), int(lmk[1])
    cv2.circle(test_image, (x, y), radius=3, color=(255, 0, 0), thickness=1)

#boundingbox

top_left = landmarks[4]
top_right = landmarks[7]
width = int(top_right[0] - top_left[0])
margin = int(width * 0.25)
height = width 
x = int(top_left[0])
y = int(top_left[1])
cv2.rectangle(test_image, (x - margin, y - margin), (x + width + margin , y + height + margin), (0, 255, 0), 2)  # Green box


plt.figure(figsize=(8,8))
plt.imshow(test_image)
plt.axis('off')
plt.show
```

```python
from torchvision.transforms import v2
```

```python
def get_transform():
    return v2.Compose([
        v2.ToImage(),
        v2.RandomHorizontalFlip(p=0.5),
        v2.RandomPerspective(distortion_scale=0.6, p=0.5),
        v2.Resize(256),
        v2.ToDtype(torch.float32, scale=True)
    ])
```

```python
from PIL import Image 
import torch 
img = Image.open(os.path.join(base_path, '00000001_012.jpg'))
```

```python
transform = get_transform()
t_img = transform(img)
```
